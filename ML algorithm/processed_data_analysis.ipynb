```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
# read final processed data and output data description
df = pd.read_csv('inputs/final_processed_data.csv')
print(df.shape)
df.head()
```

```python
# plot histograms for each feature
print("Centers:", df['center'].sum())
print("Backs:", df['back'].sum())
print("Wings:", df['wing'].sum())
print("Lines:", df['line'].sum())

plt.figure()
df['height'].hist()
plt.title('Height histogram')
plt.show()

plt.figure()
df['matches_year1'].hist()
plt.title('Matches played histogram')
plt.show()

plt.figure()
df['goals_year1'].hist()
plt.title('Goals scored histogram')
plt.show()

plt.figure()
df['goals_per_match_year1'].hist(bins = 50)
plt.title('Goals scored per match histogram')
plt.xlim(0,10)
plt.show()
```

```python
# plot correlation matrix - all players
plt.figure(figsize=[8,6])
sns.heatmap(df[["height","goals_year1","matches_year1","goals_per_match_year1",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()

plt.figure(figsize=[8,6])
sns.heatmap(df[["height","goals_year2","matches_year2","goals_per_match_year2",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```

```python
# plot correlation matrix - centers
centers = df[df['center'] == 1]
plt.figure(figsize=[8,6])
sns.heatmap(centers[["height","goals_year1","matches_year1","goals_per_match_year1",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()

plt.figure(figsize=[8,6])
sns.heatmap(centers[["height","goals_year2","matches_year2","goals_per_match_year2",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```

```python
# plot correlation matrix - backs
backs = df[df['back'] == 1]
plt.figure(figsize=[8,6])
sns.heatmap(backs[["height","goals_year1","matches_year1","goals_per_match_year1",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()

plt.figure(figsize=[8,6])
sns.heatmap(backs[["height","goals_year2","matches_year2","goals_per_match_year2",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```

```python
# plot correlation matrix - wings
wings = df[df['wing'] == 1]
plt.figure(figsize=[8,6])
sns.heatmap(wings[["height","goals_year1","matches_year1","goals_per_match_year1",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()

plt.figure(figsize=[8,6])
sns.heatmap(wings[["height","goals_year2","matches_year2","goals_per_match_year2",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```

```python
# plot correlation matrix - lines
lines = df[df['line'] == 1]
plt.figure(figsize=[8,6])
sns.heatmap(lines[["height","goals_year1","matches_year1","goals_per_match_year1",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()

plt.figure(figsize=[8,6])
sns.heatmap(lines[["height","goals_year2","matches_year2","goals_per_match_year2",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```

```python
# plot correlation matrix - position to output
plt.figure(figsize=[8,6])
sns.heatmap(df[["center","back","wing","line",
                "goals_per_match_year3"]].corr(), annot=True)
plt.show()
```
